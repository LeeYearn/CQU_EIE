```python
import os
import requests
from concurrent.futures import ThreadPoolExecutor
```

```python
# 图片保存目录
output_dir = r"C:\Users\admin\Desktop\CQU_EIE\dataEngineer\trafSignReco\img"
os.makedirs(output_dir, exist_ok=True)
```

```python
# API 基础 URL
api_url = "https://unsplash.com/napi/search/photos?page={page}&per_page=30&query=traffic+light&xp=search-region-awareness%3Acontrol"
```

```python
# 下载并保存图片
def download_image(url, path):
    with requests.get(url, stream=True) as response:
        if response.status_code == 200:
            with open(path, 'wb') as f:
                for chunk in response.iter_content(1024):
                    f.write(chunk)
            print(f"Downloaded: {path}")
```

```python
# 并发下载图片
with ThreadPoolExecutor(max_workers=10) as executor:
    for page in range(1, 501):
        response = requests.get(api_url.format(page=page)).json()
        for item in response.get("results", []):
            url = item["urls"]["full"].split("&")[0]
            path = os.path.join(output_dir, f"{item['id']}.jpg")
            executor.submit(download_image, url, path)
```
