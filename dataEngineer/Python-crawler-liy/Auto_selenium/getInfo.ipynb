```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
```

```python
# 设置 Chrome 浏览器选项
chrome_options = Options()
chrome_options.add_argument("--headless")  # 启用无头模式
chrome_options.add_argument("--disable-gpu")  # 禁用 GPU 加速
chrome_options.add_argument("--window-size=1920,1080") # 设置窗口大小
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36") # 设置用户代理
chrome_options.add_argument("--disable-extensions") # 禁用扩展
chrome_options.add_argument("--no-sandbox") # 禁用沙盒模式
chrome_options.add_argument("--disable-dev-shm-usage")  # 禁用禁用共享内存使用
```

```python
# 启动 Chrome 浏览器
service = Service('/usr/local/bin/chromedriver-linux64/chromedriver')
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
# 从 CSV 文件读取 URL
urls_df = pd.read_csv('/home/liy/cqu_eie/DataEngineer/实验 Python程序设计-爬虫的实现-liy/url.csv')  # 假设 CSV 文件有一列名为 'url'
results = []
```

```python
for index, row in urls_df.iterrows():
    # New Balance Fresh foam x 1080 v14 trainers in blue
    # window.asos.pdp.config.product = {"productCode":"139773954","name":"New Balance Fresh foam x 1080 v14 trainers in blue",

    # £160.00
    start_time = time.time()
    url = row['url']
    driver.get(url)

    # 等待页面加载
    driver.implicitly_wait(5)

    # 使用类名提取产品名称和价格
    product_name_element = driver.find_elements(By.CLASS_NAME, 'jcdpl')
    price_element = driver.find_elements(By.CLASS_NAME, 'MwTOW')

    if product_name_element and price_element:  # 检查是否找到了元素
        product_name = product_name_element[0].text.strip()
        price = price_element[0].text.strip()

        end_time = time.time()
        elapsed_time = end_time - start_time

        print(f"第{index+1}个产品名称: {product_name}")
        print(f"第{index+1}个产价格: {price}")
        print(f"第{index+1}个产品用时: {elapsed_time}秒\n")

        # 保存结果
        results.append({'产品名称': product_name, '价格': price})
    else:
        continue
```

```python
# 保存到 CSV 文件
results_df = pd.DataFrame(results)
dir = "/home/liy/cqu_eie/DataEngineer/实验 Python程序设计-爬虫的实现-liy/info.csv"
results_df.to_csv(dir, index=False, mode='a', header=not pd.io.common.file_exists(dir))
```

```python
# 关闭浏览器
driver.quit()
```
